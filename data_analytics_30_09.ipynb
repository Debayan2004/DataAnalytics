# Removing missing values like Nan

```python
# Removing rows

new_df = df.dropna()
print(new_df.to_string())
```

```python
# If you want to change the original DataFrame

df = pd.read_csv('/content/dirtydata.csv.xls')
df.dropna(inplace = True)
print(df.to_string())
```

```python
# freplace NULL values with the number 130

df = pd.read_csv('/content/dirtydata.csv.xls')
df
df.fillna(130, inplace = True)
df

```

```python
# Replace only for Specified Columns

df = pd.read_csv('/content/dirtydata.csv.xls')
df["Calories"].fillna(13, inplace = True)

df
```

```python
# Replace using Mean, Median, or Mode

df = pd.read_csv('/content/dirtydata.csv.xls')
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
df
x
```

```python
# Calculate the median and replace any empty values with it

df = pd.read_csv('/content/dirtydata.csv.xls')
x = df["Calories"].median()
df["Calories"].fillna(x, inplace = True)
df
x
```

```python
import numpy as np

# dictionary of lists

dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [np.nan, 45, 56, np.nan],
        'Third Score':[np.nan,40,80,98]}

# creating a dataframe from list

df = pd.DataFrame(dict)

# filling a missing value with
# previous ones
print(df.to_string())
df.fillna(method ='pad')
```

```python
# using isnull() function
print(df.isnull())
```

```python
# using notnull() function

print(df.notnull())
```

```python
# Padding

df.fillna(method ='pad')
```

```python
# Replace Nan with a particular value

print(df.replace(to_replace = np.nan, value = 40))
```

```python
# Replace NaN with column's mean

df_mean = df.mean()
print(df_mean)

data_new = df.fillna(df.mean())
print(data_new)
```

```python
# dropping rows if all values in that row are missing, using dropna() function

df.dropna(how = 'all')
```
